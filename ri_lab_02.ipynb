```python

import pandas as pd
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
import matplotlib.pyplot as plt

nltk.download('stopwords')
stopwords = stopwords.words('portuguese')
```

```python
df = pd.read_csv('results.csv')

ft = df['text']

tokenizer = RegexpTokenizer(r'\b[^\d\W]+')

tokens = []

for document in ft:
    tokens = tokens + tokenizer.tokenize(document.lower())

tf = []

for token in tokens:
    if (token not in stopwords and len(token) > 2):
        tf.append(token)

frequency = nltk.FreqDist(tf)
```

## 1. Reproduza a tabela 4.1 do livro texto calculando as estatísticas para sua coleção (5 pts).

```python
documents_size = len(ft)
words_size = len(tf)
vocabulary_size = len(set(tf))
ocurrency1_word = 0
ocurrency1000_word = 0


for key in frequency.keys():
    if frequency[key] == 1000:
        ocurrency1000_word += 1
    elif frequency[key] == 1:
        ocurrency1_word += 1

table41 = pd.DataFrame(data = {'Info': ['Total of Documents',
                                       'Total of Words',
                                       'Lenght of Vocabulary',
                                       'Words that occur one time',
                                       'Words that occur more than 1000 times'],
                              'Results': [documents_size,
                                          words_size,
                                          vocabulary_size,
                                          ocurrency1_word,
                                          ocurrency1000_word]})

table41
```

## 2. Reproduza a tabela 4.2 do livro considerando as top-50 palavras mais frequentes de sua coleção (5 pts).

```python
ff = pd.DataFrame(frequency.most_common(50), columns = ['word', 'frequency'])

ff['r'] = ff['frequency'].rank(ascending = False)
ff['Pr'] = ff['frequency'] / len(tokens)
ff['Pr %'] = (ff['frequency'] / len(tokens) * 100)
ff['r.Pr'] = ff['r'] * (ff['frequency'] / len(tokens))


ff
```

## 3. Plote curvas de rank-frequência (usando escala logarítmica) para palavras e bigramas na coleção que você coletou na atividade anterior. Quais os melhores valores do parâmetro c para cada curva (5 pts)?

```python
diagrams = []
diagrams = nltk.bigrams(tf)

fd = nltk.FreqDist(diagrams)


diagrams_df = pd.DataFrame(fd.most_common(), columns=['word', 'frequency'])
diagrams_df['r'] = diagrams_df['frequency'].rank(ascending = False)
```

```python
f, ax = plt.subplots(figsize=(15, 7))

ax.set(xscale='log', yscale='log')


sns.regplot('r', 'frequency', diagrams_df, ax=ax, ci=68, scatter_kws={'s': 150}, color = 'g')
```

## 4. Reproduza a tabela 4.5 do livro texto considerando a sua coleção (5 pts).

```python
count = 0
result = []
for n in range(1,11):
    for j in frequency.items():
        if j[1] == n:
            count += 1
    tupple = (n, count)
    result.append(tupple)
    count = 0

    
df = pd.DataFrame(result, columns = ['Number of occurrences', 'Number of words'])
df['Planned proportion'] = (1 / (df['Number of occurrences'] * (df['Number of occurrences'] + 1)))
df['Real proportion'] = (df['Number of words'] / words_size)
df
```
